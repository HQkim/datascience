```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips') # seaborn 라이브러리에 기본적으로 제공하는 tips 데이터 셋
tips.head(10)
```

```python
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")

x = iris.petal_length.values
plt.rc("font", family = "Malgun Gothic")
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
iris.head(15)
```

```python
sns.kdeplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

```python
sns.distplot(x, kde=True, rug=True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 DIst Plot")
plt.show()
```

```python
sns.regplot(x=iris["sepal_length"], y=iris["petal_length"], scatter_kws={'color': 'orange'}, line_kws={"color" : 'green'})
plt.show()
```

```python
ax =sns.regplot(x=iris["sepal_length"], y=iris["petal_length"], scatter_kws={'color': 'orange'}, line_kws={"color" : 'green'}, fit_reg=False)
plt.show()
```

```python
ax.set_xlabel("꽃잎 길이")
ax.set_ylabel("꽃받침 길이")
ax.figure
```

```python
# Google color picker 검색해서 컬러 16진수값 color에 입력해도됨
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, color ='#dade10')
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot과 Kernel Density Plot", y=1.02)
plt.show()
```

```python
sns.pairplot(iris)
plt.title("Iris Data의 Pair Plot",x=-1.2, y =3.3, fontsize=20)
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o","s","D"])
plt.title("Iris Pair Plot, Hue로 꽃의 종을 시각화",x=-1.2, y =3.3, fontsize=20)
plt.show()
```

```python
covid = df[['연번','확진일','지역']]
print(covid['확진일'].value_counts())
```

```python
print(covid['확진일'].head(1))
covid['확진일']= pd.to_datetime("2020"+"-"+covid["확진일"].str.replace(".","-"))
print(covid['확진일'])
print(covid['확진일'].head())
```

```python
covid['월']=covid['확진일'].dt.month
print(covid['월'])
print(covid)
```

```python
covid['주'] = covid['확진일'].dt.week
covid['월일']=covid['확진일'].astype(str).map(lambda x:x[-5:])
covid
```

```python
import matplotlib.pyplot as plt
```

```python
covid['확진일'].value_counts().sort_index().plot(figsize=(15,4),rot=50,grid=True)
plt.axhline(30, ls=":", color="red")
plt.show()
```

```python
day_count = covid['확진일'].value_counts().sort_index()
print(day_count.describe())
```

```python
print(day_count[day_count == day_count.max()])
print(covid[covid["확진일"] == day_count[day_count == day_count.max()].index[0]]
)
```

```python
g = day_count.plot.bar(figsize=(30,10))
g.axhline(30,ls=":")
plt.show()
```
