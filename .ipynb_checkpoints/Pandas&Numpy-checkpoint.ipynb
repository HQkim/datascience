```python
# 내 답안
term1 = (company['업종'] == '기타 금융업')
term2 = (company['분류'] == '코스닥')
term3 = (pd.to_datetime(company['상장일']) > pd.to_datetime('2015-1-1'))
# term3 = (pd.to_datetime(company['상장일']) > '2015-1-1') # str으로 비교됨
# term3 = (company['상장일'] > '2015-1-1')
company[ term1 & term2 & term3 ]

```

```python
# 자체적으로 데이터를 뽑을때 날짜를 Timestamp로 바꿈
# company = pd.read_csv('company.csv', parse_dates=['상장일'])
```

```python
financial_corp = company[company['업종'] == '기타 금융업']
kosdaq_financial_corp = financial_corp[financial_corp['분류'] == '코스닥']
kosdaq_financial_corp['상장일'] = pd.to_datetime(kosdaq_financial_corp['상장일'])
kosdaq_financial_corp[kosdaq_financial_corp['상장일'] > pd.Timestamp(2015, 1, 1)] # '2015-1-1'로 해도 되나 나중에 연도만 따로 뽑거나 하기 힘들다. 문자로 비교하게됨.
# deprecitated: 지금은 지원하지만 조만간 바꿀꺼다.
# 아래 문법이 낫다. 하지만 무시해도 돌간다.
```
