# 1.Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 2. Reda DataSet

### 2-1. Feature Data(X)

```python
feature_df = pd.read_excel('boston_house_data.xlsx', index_col=0)
print(feature_df.shape)
feature_df.head()
```

### 2.1 Target Data

```python
target_df = pd.read_excel('boston_house_target.xlsx', index_col=0)
print(target_df.shape)
target_df.head()
```

### 2-3. Concat Feature & Target Df

```python
total_df = pd.concat([feature_df, target_df], axis=1)
print(total_df.shape)
total_df.head()
```

```python
total_df.columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
total_df.head()
```

```python
total_df.describe()
```

# 3. Convert DataFrame to Numpy array

```python
feature_array = np.array(feature_df)
target_array = np.array(target_df)
print(feature_array.ndim)
print(feature_array.shape)
```

```python
print(target_array.ndim)
print(target_array.shape)
```

# 4. Split into Training & Test set

```python
feature_num = 12
```

```python
total_df
```

```python
boston_x = feature_array[:,feature_num:feature_num+1] # boston_x = feature_array[:,1:5] 같이 주면 다중회기분석
print(boston_x)
```

```python
boston_y = target_array
print(target_array)
```

```python
from sklearn import model_selection

x_train, x_test, y_train, y_test = model_selection.train_test_split(boston_x,boston_y,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
boston_x.shape
```

# 5. Create Model object

```python
from sklearn import linear_model

model = linear_model.LinearRegression() # 클래스를 호출
model
```

# 6. Train Model object

```python
model.fit(x_train, y_train)
```

```python
print('Coefficeint: ',model.coef_) # theta 1
```

```python
print('Intercept: ', model.intercept_) # theta 0
```

```python
print('y = {:.4f}x + {:.4f}'.format(model.coef_[0][0], model.intercept_[0]))
```

# 7. Test Model Object

### 7-1. Mean Squared Error with Pure Python

```python
# model.predict(x_train)
```

```python
print('MSE(training data): ',np.mean((model.predict(x_train) - y_train)**2))
```

```python
print('MSE(test data): ',np.mean((model.predict(x_test) - y_test)**2))
```

### 7-2 MSE with Sklearn

```python
from sklearn.metrics import mean_squared_error

print('MSE(test data):', mean_squared_error(model.predict(x_test), y_test)) # 끝자리 까지 같다.
```

# 8. Visualization

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(x_test, y_test, color="black")
plt.scatter(x_train, y_train, color="red", s=1)
plt.plot(x_test, model.predict(x_test), color="blue", linewidth=3)
plt.plot(x_train, model.predict(x_train), color="yellow", linewidth=1) # model은 이미 학습 완료되어 parameter 값 고정이기에 train, test 둘중 아무거나 그리면 된다. 

plt.grid()
plt.show()
```
